# Part 1: Basic Linear Regression

This section covers the fundamental building blocks of linear regression using intuitive concepts and basic algebra. **No calculus or matrix derivations are needed here** — those are covered in Part 2.

---

<a name="section_1_1"></a>
### 1.1 The Linear Hypothesis

A linear regression model predicts an output $\hat{y}$ from an input $x$ using a straight line. There are two common forms:

**1-Parameter model** (line forced through the origin, no intercept):
$$\boxed{\hat{y} = \theta_1 \, x} \tag{1.1}$$

> **Numerical Example 1.1 (using Eq. 1.1):**
> Let $\theta_1 = 2.0$ and $x = 3.0$:
> $$\hat{y} = 2.0 \times 3.0 = 6.0$$

**2-Parameter model** (the standard form with intercept):
$$\boxed{\hat{y} = \theta_0 + \theta_1 \, x} \tag{1.2}$$

* $\theta_0$ — the **intercept**: value of $\hat{y}$ when $x = 0$.
* $\theta_1$ — the **slope**: how much $\hat{y}$ changes per unit increase in $x$.

> **Numerical Example 1.2 (using Eq. 1.2):**
> Let $\theta_0 = 1.0$, $\theta_1 = 1.5$, and $x = 2.0$:
> $$\hat{y} = 1.0 + 1.5 \times 2.0 = 1.0 + 3.0 = 4.0$$

**Multi-feature extension** (features $x_1, x_2, \ldots, x_d$):

**Compact (Vector) Version:**
$$\boxed{\hat{y} = \theta^{T} x} \tag{1.3}$$

**Expanded Version:**
$$\hat{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \cdots + \theta_d x_d$$
*(where we prepend a dummy feature $x_0 = 1$ to the input vector $x = [1, x_1, x_2, \ldots, x_d]^T$ and parameter vector $\theta = [\theta_0, \theta_1, \theta_2, \ldots, \theta_d]^T$)*

> **Numerical Example 1.3 (using Eq. 1.3):**
> Let $\theta_0 = 10.0$, $\theta_1 = 2.5$, $\theta_2 = -1.2$, with features $x_1 = 4.0$, $x_2 = 2.0$:
> $$\hat{y} = 10.0 + (2.5 \times 4.0) + (-1.2 \times 2.0) = 10.0 + 10.0 - 2.4 = 17.6$$


<a name="section_1_2"></a>
### 1.2 Loss Function (Residuals)

After making a prediction $\hat{y}$, we compare it to the actual value $y$. The **residual** is the difference:

$$\boxed{e^{(i)} = y^{(i)} - \hat{y}^{(i)}} \tag{1.4}$$

Because residuals can be positive or negative, we **square** them so that errors always add up (never cancel). The **loss** for one sample is:

$$\boxed{\text{Loss}^{(i)} = \left( \hat{y}^{(i)} - y^{(i)} \right)^2} \tag{1.5}$$

A perfect prediction gives Loss $= 0$. The larger the error, the larger the loss.

> **Numerical Example 1.4 (using Eqs. 1.4–1.5, with $\hat{y} = 4.0$):**
>
> | $y$ | Residual $$(y - \hat{y})$$ | Loss $$(\hat{y}-y)^2$$ |
> |:---:|:---:|:---:|
> | 3.0 | $$3.0 - 4.0 = -1.0$$ | $$(-1.0)^2 = 1.00$$ |
> | 4.0 | $$4.0 - 4.0 = 0.0$$ | $$(0.0)^2 = 0.00$$ |
> | 4.5 | $$4.5 - 4.0 = +0.5$$ | $$(0.5)^2 = 0.25$$ |

👉 *For a visual of residuals on a regression plot, see [Section 3.1](#section_3_1).*


<a name="section_1_3"></a>
### 1.3 Cost Function (Mean Squared Error)

The **loss** (Eq. 1.5) measures error for one sample. The **cost function** $J$ averages the loss across *all* $m$ training samples, giving a single number that summarises overall model fit:

**Compact Version:**
$$\boxed{J(\theta_0, \theta_1) = \frac{1}{2m} \sum_{i=1}^{m} (e^{(i)})^2} \tag{1.6}$$

**Expanded Version** (by substituting the 2-parameter hypothesis Eq. 1.2):
$$J(\theta_0, \theta_1) = \frac{1}{2m} \sum_{i=1}^{m} \left( (\theta_0 + \theta_1 x^{(i)}) - y^{(i)} \right)^2$$

* The $\frac{1}{2}$ factor is a mathematical convenience (it simplifies the derivative — see Part 2).
* **Our goal:** find $\theta_0$ and $\theta_1$ that make $J$ as small as possible.

> **Numerical Example 1.5 (using Eq. 1.6, dataset: $x=[1,2,3]$, $y=[2,4,6]$, $\theta_0=1.0$, $\theta_1=1.5$):**
>
> | $i$ | $x$ | $y$ | $\hat{y}$ $$(\theta_0{=}1,\, \theta_1{=}1.5)$$ | $$(\hat{y}-y)^2$$ |
> |:---:|:---:|:---:|:---:|:---:|
> | 1 | 1 | 2 | $$2.5$$ | $$0.25$$ |
> | 2 | 2 | 4 | $$4.0$$ | $$0.00$$ |
> | 3 | 3 | 6 | $$5.5$$ | $$0.25$$ |
>
> $$J(1.0,\, 1.5) = \frac{1}{2 \times 3}(0.25 + 0.00 + 0.25) = \frac{0.50}{6} \approx 0.0833$$


<a name="section_1_4"></a>
### 1.4 Optimization: Finding the Best Parameters

Our goal is to find $\theta_0$ and $\theta_1$ that minimise the cost $J$ (Eq. 1.6). There are two main approaches:

---

#### A. Normal Equation (Direct / Closed-Form)

The **Normal Equation** gives the exact optimal parameters in **one calculation** using linear algebra:

$$\boxed{\theta = (X^{T} X)^{-1} X^{T} y} \tag{1.7}$$

Here $X$ is the **design matrix** — a table where each row is one training sample and the first column is all $1$s (for the intercept). Think of Eq. 1.7 as solving a system of equations directly — no guessing or iterating required.

> **When to use:** Works well for small-to-medium datasets (up to ~10,000 features).

> **Numerical Example 1.6 (2-Parameter Model using Eq. 1.7, dataset: $x=[1,2,3]$, $y=[2,4,6]$):**
>
> **Step 1** — Build the design matrix $X$ (add a column of $1$s for $\theta_0$):
> $$X = \begin{bmatrix} 1 & 1 \\ 1 & 2 \\ 1 & 3 \end{bmatrix}, \quad y = \begin{bmatrix} 2 \\ 4 \\ 6 \end{bmatrix}$$
>
> **Step 2** — Compute $X^T X$ and $X^T y$:
> $$X^T X = \begin{bmatrix} 3 & 6 \\ 6 & 14 \end{bmatrix}, \qquad X^T y = \begin{bmatrix} 12 \\ 28 \end{bmatrix}$$
>
> **Step 3** — Invert $X^T X$ (determinant $= 3 \times 14 - 6 \times 6 = 6$):
> $$(X^T X)^{-1} = \frac{1}{6}\begin{bmatrix} 14 & -6 \\ -6 & 3 \end{bmatrix} = \begin{bmatrix} 2.333 & -1.0 \\ -1.0 & 0.5 \end{bmatrix}$$
>
> **Step 4** — Multiply to get $\theta$:
> $$\theta = \begin{bmatrix} 2.333 & -1.0 \\ -1.0 & 0.5 \end{bmatrix} \begin{bmatrix} 12 \\ 28 \end{bmatrix} = \begin{bmatrix} (2.333 \times 12) - (1.0 \times 28) \\ -(1.0 \times 12) + (0.5 \times 28) \end{bmatrix} = \begin{bmatrix} 0.0 \\ 2.0 \end{bmatrix}$$
>
> **Result:** $\theta_0 = 0.0$, $\theta_1 = 2.0$ — a perfect fit!

👉 *For the full matrix calculus derivation of Eq. 1.7, see [Section 2.1](#section_2_1).*

---

#### B. Gradient Descent (Iterative)

**Gradient Descent** finds the minimum by starting with an initial guess and taking repeated small steps downhill on the cost surface, until the cost stops decreasing.

**Intuition:** Imagine standing on a hilly landscape and always stepping in the steepest downhill direction. Eventually you reach the valley (minimum cost).

**Conceptual (Compact) Update Rule** (applied simultaneously to all parameters):
$$\boxed{\theta_j := \theta_j - \alpha \cdot \text{Slope}_j} \tag{1.8}$$

* $\alpha$ is the **learning rate** — controls step size.
  * Too large → overshoots the minimum.
  * Too small → converges very slowly.
* $\text{Slope}_j$ represents the direction and steepness of the hill for parameter $\theta_j$.

**Expanded Update Rules for the 2-Parameter Model** ($\hat{y} = \theta_0 + \theta_1 x$):
$$\theta_0 := \theta_0 - \alpha \cdot \text{Slope}_0$$
$$\theta_1 := \theta_1 - \alpha \cdot \text{Slope}_1$$
*(where the exact mathematical formulas for calculating $\text{Slope}_0$ and $\text{Slope}_1$ are derived using calculus in Section 2.2 — see Eqs. 2.11 and 2.12)*

> **Numerical Example 1.7 (using Eq. 1.8):**
> Suppose we are optimizing parameter $\theta_1$. Its current value is $\theta_1 = 1.0$. If we set the learning rate $\alpha = 0.1$, and the calculated downhill slope is $\text{Slope}_1 = -4.0$:
> $$\theta_1 := 1.0 - 0.1 \times (-4.0) = 1.0 + 0.4 = 1.4$$

> **Convergence Snapshot** (starting at $\theta_0=0$, $\theta_1=0$, learning rate $\alpha=0.1$):
>
> | Step | $\theta_0$ | $\theta_1$ | Cost $J$ |
> |:---:|:---:|:---:|:---:|
> | 0 | 0.000 | 0.000 | 9.333 |
> | 1 | 0.400 | 0.933 | 1.816 |
> | 2 | 0.573 | 1.351 | 0.354 |
> | 5 | 0.120 | 1.890 | 0.010 |
> | 10 | 0.003 | 1.997 | 0.000 |
>
> After 10 steps the parameters converge to $\approx [0.003,\; 1.997]$, very close to the true $[0.0,\; 2.0]$.

👉 *For the partial derivative formulas behind Eq. 1.8, see [Section 2.2](#section_2_2).*

---

#### Practical Comparison: Gradient Descent vs. Normal Equation

| Feature | Gradient Descent | Normal Equation |
| :--- | :--- | :--- |
| **Method** | Iterative (many steps) | Analytical (one-shot, Eq. 1.7) |
| **Learning rate $\alpha$** | Must be chosen carefully | Not needed |
| **Speed — many features** | Fast — scales well to millions | Slow — matrix inversion is $O(d^3)$ |
| **Speed — few features** | Slower (needs many steps) | Fast and exact |
| **Feature scaling** | Recommended for faster convergence | Not required |
| **Non-invertible $X^T X$** | Not an issue | Can fail (use pseudo-inverse) |

### 1.5 Train-Test Split (Generalization)

A model must perform well on **new, unseen data**, not just the data it was trained on. This ability is called **generalisation**.

To measure it, we split the dataset *before* training:
* **Training set** (~70%): used to fit the model parameters.
* **Test set** (~30%): kept hidden during training; used only for final evaluation.

If a model performs well on training data but poorly on the test set, it has **overfit** — it memorised the training noise.

> **Example:** Dataset of 100 house price records.
> * Train on 70 → fit $\theta_0$ and $\theta_1$ (Eq. 1.2).
> * Test on the remaining 30 → report real-world performance.

💻 *For code implementation, see [Section 3.3](#section_3_3).*


<a name="section_1_6"></a>
### 1.6 Cross-Validation (Model Validation)

A single train-test split can give a biased result depending on *which* samples land in each set. **K-Fold Cross-Validation** provides a more reliable estimate:

1. Divide the dataset into $K$ equal chunks (folds), e.g., $K = 5$.
2. Run $K$ experiments. In experiment $k$, use fold $k$ as the test set and the remaining $K-1$ folds for training.
3. Average the $K$ performance scores to get a stable, unbiased estimate.

> **Example with $K = 5$ on 100 samples:**
> Each fold has 20 samples. We train 5 times, each time testing on a different 20-sample block.
> Final score = average of the 5 individual test scores.

This ensures every sample is tested exactly once, giving a more reliable evaluation than any single split.

### 1.7 Regularization (Overfitting, Ridge & Lasso Concepts)

#### What is Overfitting?

When a model is too complex, it fits the training data nearly perfectly but fails on new data — it has memorised noise instead of learning the true pattern.

**Signs of overfitting:**
* Very low training cost (Eq. 1.6), but high test cost.
* Model predictions are overly sensitive to small input changes.

#### Regularization: Penalising Complexity

Regularization adds a **penalty term** to the cost function that discourages large parameter values:

$$\boxed{J_{\text{regularized}}(\theta) = J_{\text{MSE}}(\theta) + \lambda \cdot \text{Penalty}(\theta)} \tag{1.9}$$

*(Here $J_{\text{MSE}}$ is the standard cost from Eq. 1.6, and $\lambda \cdot \text{Penalty}(\theta)$ discourages overly large parameters.)*

* $\lambda \ge 0$ controls the strength of the penalty.
  * $\lambda = 0$ → no penalty (standard linear regression).
  * Large $\lambda$ → stronger penalty → parameters shrink toward zero.

**Two common types:**

**Ridge Regression (L2)**:
**Compact Version:**
$$\boxed{J_{\text{Ridge}} = J_{\text{MSE}} + \lambda \sum_{j=1}^{d} \theta_j^2} \tag{1.10}$$
**Expanded Version** (for a 2-parameter model, where the intercept $\theta_0$ is conventionally not penalised):
$$J_{\text{Ridge}} = \frac{1}{2m}\sum_{i=1}^{m}\left((\theta_0 + \theta_1 x^{(i)}) - y^{(i)}\right)^2 + \lambda \theta_1^2$$

**Lasso Regression (L1)**:
**Compact Version:**
$$\boxed{J_{\text{Lasso}} = J_{\text{MSE}} + \lambda \sum_{j=1}^{d} |\theta_j|} \tag{1.11}$$
**Expanded Version** (for a 2-parameter model, where the intercept $\theta_0$ is conventionally not penalised):
$$J_{\text{Lasso}} = \frac{1}{2m}\sum_{i=1}^{m}\left((\theta_0 + \theta_1 x^{(i)}) - y^{(i)}\right)^2 + \lambda |\theta_1|$$

| | **Ridge (L2)** | **Lasso (L1)** |
|:---|:---:|:---:|
| **Penalty** | $$\lambda \sum \theta_j^2$$ | $$\lambda \sum |\theta_j|$$ |
| **Shrinks parameters?** | Yes, toward 0 | Yes, can reach exactly 0 |
| **Feature selection?** | No | **Yes** — removes irrelevant features |

> **Numerical Example 1.8 (using Eqs. 1.10–1.11, $J_{\text{MSE}}=0.0833$, $\theta_1=1.5$, $\lambda=0.2$):**
> * **Ridge:** $0.0833 + 0.2 \times (1.5)^2 = 0.0833 + 0.45 = 0.5333$
> * **Lasso:** $0.0833 + 0.2 \times |1.5| = 0.0833 + 0.30 = 0.3833$

### 2.1 Derivation of the Normal Equation (Matrix Calculus)

**Setup:** Let $X \in \mathbb{R}^{m \times (d+1)}$ be the design matrix (first column = all $1$s), $y \in \mathbb{R}^{m}$ the target vector, $\theta \in \mathbb{R}^{d+1}$ the parameter vector.

**Step 1 — Write the cost in matrix form:**
$$J(\theta) = \frac{1}{2m} (X\theta - y)^{T} (X\theta - y) \tag{2.1}$$

**Step 2 — Expand the transpose product:**
$$(X\theta - y)^{T}(X\theta - y) = \theta^{T} X^{T} X \theta - 2\theta^{T} X^{T} y + y^{T} y \tag{2.2}$$

> *Note: $\theta^T X^T y$ is a scalar, so it equals its own transpose $y^T X \theta$, allowing us to combine the two cross terms into $-2\theta^T X^T y$.*

Substituting back:
$$J(\theta) = \frac{1}{2m} \left( \theta^{T} X^{T} X \theta - 2\theta^{T} X^{T} y + y^{T} y \right) \tag{2.3}$$

**Step 3 — Differentiate with respect to $\theta$ and set to zero:**

Using matrix calculus rules [$\dots$]:

$$\nabla_{\theta} J(\theta) = \frac{1}{2m} \left( 2 X^{T} X \theta - 2 X^{T} y \right) = \frac{1}{m} \left( X^{T} X \theta - X^{T} y \right) = 0 \tag{2.4}$$

**Step 4 — Solve for $\theta$:**
$$X^{T} X \theta = X^{T} y \tag{2.5}$$

Multiply both sides by $(X^T X)^{-1}$ (assuming it is invertible):
$$\boxed{\theta = (X^{T} X)^{-1} X^{T} y} \tag{2.6}$$

This is the same as Eq. 1.7 — now formally derived. If $X^T X$ is not invertible, a pseudo-inverse or regularization is needed.

### 2.2 Derivation of Gradient Descent Gradients (Chain Rule)

Gradient descent uses Eq. 1.8 to update each parameter:
$$\theta_{j} := \theta_{j} - \alpha \frac{\partial J}{\partial \theta_{j}} \tag{2.7}$$

We now derive the partial derivative $\frac{\partial J}{\partial \theta_j}$ formally:

$$\frac{\partial J}{\partial \theta_{j}} = \frac{\partial}{\partial \theta_{j}} \left( \frac{1}{2m} \sum_{i=1}^{m} \left( h_{\theta}(x^{(i)}) - y^{(i)} \right)^{2} \right) \tag{2.8}$$

Apply the **power rule** (outer derivative) and **chain rule** (inner derivative):

$$= \frac{1}{m} \sum_{i=1}^{m} \left( h_{\theta}(x^{(i)}) - y^{(i)} \right) \cdot \frac{\partial h_{\theta}(x^{(i)})}{\partial \theta_{j}} \tag{2.9}$$

Since $h_\theta(x^{(i)}) = \sum_k \theta_k x_k^{(i)}$, its derivative with respect to $\theta_j$ is $x_j^{(i)}$. Therefore:

$$\boxed{\frac{\partial J}{\partial \theta_{j}} = \frac{1}{m} \sum_{i=1}^{m} \left( h_{\theta}(x^{(i)}) - y^{(i)} \right) x_{j}^{(i)}} \tag{2.10}$$

**For the 2-parameter model** $h(x) = \theta_0 + \theta_1 x$ (where $x_0^{(i)} = 1$):

$$\frac{\partial J}{\partial \theta_{0}} = \frac{1}{m} \sum_{i=1}^{m} \left( h_{\theta}(x^{(i)}) - y^{(i)} \right) \tag{2.11}$$

$$\frac{\partial J}{\partial \theta_{1}} = \frac{1}{m} \sum_{i=1}^{m} \left( h_{\theta}(x^{(i)}) - y^{(i)} \right) x^{(i)} \tag{2.12}$$

**Vectorized update rule** (efficient matrix form for implementation):
$$\boxed{\theta := \theta - \frac{\alpha}{m} X^{T} (X\theta - y)} \tag{2.13}$$

### 2.3 Convergence Math for 2-Parameter Gradient Descent

**Setup:** $x=[1,2,3]$, $y=[2,4,6]$, $m=3$, $\theta_0^{(0)}=0$, $\theta_1^{(0)}=0$, $\alpha=0.1$.

We apply Eqs. 2.11–2.12 and 2.7 manually for two iterations.

---

#### Iteration 1

1. **Predictions** (using Eq. 1.2): $h(x) = 0 + 0 \cdot x = [0,\; 0,\; 0]$
2. **Errors** $(h(x^{(i)}) - y^{(i)})$: $[-2,\; -4,\; -6]$
3. **Gradients** (Eqs. 2.11–2.12):
   $$\frac{\partial J}{\partial \theta_0} = \frac{-2 + (-4) + (-6)}{3} = -4.000$$
   $$\frac{\partial J}{\partial \theta_1} = \frac{(-2)(1) + (-4)(2) + (-6)(3)}{3} = \frac{-28}{3} \approx -9.333$$
4. **Updates** (Eq. 2.7, $\alpha=0.1$):
   $$\theta_0^{(1)} = 0 - 0.1 \times (-4.000) = 0.400$$
   $$\theta_1^{(1)} = 0 - 0.1 \times (-9.333) = 0.933$$

---

#### Iteration 2

1. **Predictions:** $h(x) = 0.4 + 0.933x = [1.333,\; 2.267,\; 3.200]$
2. **Errors:** $[-0.667,\; -1.733,\; -2.800]$
3. **Gradients** (Eqs. 2.11–2.12):
   $$\frac{\partial J}{\partial \theta_0} = \frac{-0.667 + (-1.733) + (-2.800)}{3} \approx -1.733$$
   $$\frac{\partial J}{\partial \theta_1} = \frac{(-0.667)(1) + (-1.733)(2) + (-2.800)(3)}{3} \approx -4.178$$
4. **Updates** (Eq. 2.7):
   $$\theta_0^{(2)} = 0.400 - 0.1 \times (-1.733) = 0.573$$
   $$\theta_1^{(2)} = 0.933 - 0.1 \times (-4.178) = 1.351$$

---

After 2 iterations, $[\theta_0, \theta_1]$ moves from $[0, 0]$ toward the true solution $[0, 2]$. The code in [Section 4.2](#section_4_2) shows full convergence over 10 steps.


<a name="section_2_4"></a>
### 2.4 Mathematical Formulation of Regularization (Ridge & Lasso)

#### A. Ridge Regression (L2 Regularization)

$$\boxed{J_{\text{Ridge}}(\theta) = \frac{1}{2m} \sum_{i=1}^{m} \left(h_{\theta}(x^{(i)}) - y^{(i)}\right)^{2} + \lambda \sum_{j=1}^{d} \theta_{j}^{2}} \tag{2.14}$$

This is the full form of Eq. 1.10. The penalty $\sum_{j=1}^{d} \theta_j^2$ is the squared **L2 norm** $\|\theta\|_2^2$. The intercept $\theta_0$ is conventionally **not** penalised.

Taking the gradient of Eq. 2.14 with respect to $\theta_j$ ($j \ge 1$):

$$\frac{\partial J_{\text{Ridge}}}{\partial \theta_j} = \frac{1}{m}\sum_{i=1}^{m}\left(h_\theta(x^{(i)}) - y^{(i)}\right)x_j^{(i)} + 2\lambda\theta_j \tag{2.15}$$

#### B. Lasso Regression (L1 Regularization)

$$\boxed{J_{\text{Lasso}}(\theta) = \frac{1}{2m} \sum_{i=1}^{m} \left(h_{\theta}(x^{(i)}) - y^{(i)}\right)^{2} + \lambda \sum_{j=1}^{d} |\theta_{j}|} \tag{2.16}$$

This is the full form of Eq. 1.11. The penalty $\sum_{j=1}^{d}|\theta_j|$ is the **L1 norm** $\|\theta\|_1$. Since $|\theta_j|$ is not differentiable at $\theta_j = 0$, Lasso uses **subgradient** or **coordinate descent** methods.

#### Why Lasso Produces Sparse Solutions

The constraint regions explain why Lasso can drive coefficients to exactly zero:
* **Ridge (L2):** Constraint region is a smooth sphere $\sum \theta_j^2 \le t$. The cost contours usually meet this smooth boundary at a non-zero point.
* **Lasso (L1):** Constraint region is a diamond/polytope $\sum |\theta_j| \le t$ with sharp corners on the coordinate axes. Cost contours are highly likely to meet the boundary at a **corner** where one or more $\theta_j = 0$ exactly.

This is why Lasso automatically performs **feature selection**: irrelevant features get coefficients set to exactly zero.

### 3.1 Visualizing Residual Errors (Loss)

The plot below shows the fitted regression line and **vertical lines representing residuals** (Eq. 1.4) — the individual loss for each training point. Red dashed lines highlight two specific residuals: a large positive one (x=20) and a large negative one (x=75).


```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Manually structured data points
x = np.array([6, 8, 11, 14, 17, 20, 24, 25, 28, 30, 33, 38, 40, 47, 47, 52, 53, 56, 61, 64, 66, 68, 75, 78, 84, 89])
y = np.array([1500, 2500, 3000, 1900, 2600, 4000, 2300, 3200, 1900, 3950, 3250, 2850, 3850, 2850, 4100, 2450, 3050, 3950, 2980, 4700, 3500, 4150, 2300, 4350, 3700, 4600])

# 2. Calculate linear regression (y = mx + b)
m, b = np.polyfit(x, y, 1)

x_line = np.linspace(0, 100, 100)
y_line = m * x_line + b
y_pred = m * x + b

# 3. Build the plot
plt.figure(figsize=(8, 5))
plt.plot(x_line, y_line, color='black', linewidth=1.5)

for xi, yi, y_pi in zip(x, y, y_pred):
    if xi == 20 or xi == 75:
        plt.vlines(xi, yi, y_pi, colors='red', linestyles='dashed', linewidth=2)
    else:
        plt.vlines(xi, yi, y_pi, colors='gray', alpha=0.5, linewidth=1)

plt.scatter(x, y, color='black', zorder=5)

plt.annotate('Residual', xy=(75, 3200), xytext=(77, 2800), fontsize=14, color='black',
             arrowprops=dict(facecolor='dodgerblue', shrink=0.05, width=2, headwidth=8))
plt.annotate('', xy=(20, 3300), xytext=(15, 3300),
             arrowprops=dict(facecolor='dodgerblue', shrink=0.05, width=2, headwidth=8))

plt.xlim(0, 100)
plt.ylim(0, 5000)
plt.title("Visualizing Residuals (Loss) in Linear Regression")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, alpha=0.3)
plt.show()
```

### 3.2 Scikit-Learn Linear Regression (Single Feature)

Here we fit a linear regression model (Eq. 1.2) using `scikit-learn` to predict Life Expectancy from Fertility rates (Gapminder dataset).


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('https://raw.githubusercontent.com/tofighi/MachineLearning/master/datasets/gapminder.csv')

# Select only numeric columns for correlation heatmap
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), square=True, cmap='RdYlGn')
plt.show()

# Setup features and target variable
y = df['life']
X_fertility = df[['fertility']]

# Create prediction space
prediction_space = np.linspace(min(df['fertility']), max(df['fertility'])).reshape(-1, 1)

# Fit model
reg = LinearRegression()
reg.fit(X_fertility, y)

# Predict wrapping in DataFrame to avoid feature name warning
y_pred = reg.predict(pd.DataFrame(prediction_space, columns=['fertility']))

print("R^2 score:", reg.score(X_fertility, y))
print("Intercept (theta_0):", reg.intercept_)
print("Coefficient (theta_1):", reg.coef_)

plt.scatter(X_fertility, y, color='blue', alpha=0.5)
plt.plot(prediction_space, y_pred, color='black', linewidth=3)
plt.xlabel('Fertility')
plt.ylabel('Life Expectancy')
plt.title('Predicting Life Expectancy from Fertility (Eq. 1.2)')
plt.grid(True)
plt.show()
```

### 3.3 Train-Test Split and Multi-Feature Regression (Scikit-Learn)

Below we evaluate linear regression (Eq. 1.3) using all available features, with a 70%/30% train-test split.


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

df = pd.read_csv('https://raw.githubusercontent.com/tofighi/MachineLearning/master/datasets/gapminder.csv')

X = df.drop("life", axis=1)
y = df["life"]

# One-hot encode the categorical 'Region' column
X = pd.get_dummies(X, columns=['Region'], prefix=['Region'], dtype=int)

# Train-test split (30% test, 70% train)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

reg_all = LinearRegression()
reg_all.fit(X_train, y_train)
y_pred = reg_all.predict(X_test)

print("R^2 Score:", r2_score(y_test, y_pred))
print("MAE:      ", mean_absolute_error(y_test, y_pred))
print("MSE:      ", mean_squared_error(y_test, y_pred))
print("RMSE:     ", np.sqrt(mean_squared_error(y_test, y_pred)))
```

### 4.1 Implementing Normal Equation from Scratch (Eq. 2.6)

We implement $\theta = (X^T X)^{-1} X^T y$ for both 1-parameter and 2-parameter models.


```python
import numpy as np

# Dataset
X_val = np.array([1.0, 2.0, 3.0])
y_val = np.array([2.0, 4.0, 6.0])

# --- 1-Parameter Normal Equation: y = theta_1 * x (Eq. 1.1) ---
# theta_1 = sum(x*y) / sum(x^2)  [scalar form of Eq. 2.6]
theta_1_1param = np.sum(X_val * y_val) / np.sum(X_val ** 2)
print("--- 1-Parameter Normal Equation (No Intercept, Eq. 1.1) ---")
print(f"Optimal theta_1: {theta_1_1param:.4f}\n")

# --- 2-Parameter Normal Equation: y = theta_0 + theta_1*x (Eq. 2.6) ---
# Design matrix: prepend a column of ones for the intercept term
X_design = np.vstack((np.ones(len(X_val)), X_val)).T
theta_2param = np.linalg.inv(X_design.T @ X_design) @ X_design.T @ y_val

print("--- 2-Parameter Normal Equation (Intercept + Slope, Eq. 2.6) ---")
print("Design Matrix X:\n", X_design)
print(f"Optimal theta_0 (Intercept): {theta_2param[0]:.4f}")
print(f"Optimal theta_1 (Slope):     {theta_2param[1]:.4f}")
```

<a name="section_4_2"></a>
### 4.2 Implementing Gradient Descent from Scratch (Eqs. 2.10–2.13)

We implement the iterative Gradient Descent update (Eq. 2.7 / vectorized Eq. 2.13) for both 1-parameter and 2-parameter models, and plot convergence.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
X_val = np.array([1.0, 2.0, 3.0])
y_val = np.array([2.0, 4.0, 6.0])
m = len(X_val)

# --- A. 1-Parameter GD: y = theta_1 * x (Eq. 1.1) ---
# Gradient: dJ/d_theta1 = (1/m) * sum((y_pred - y) * x)  [Eq. 2.10 with x_j = x]
theta_1 = 0.0
alpha = 0.1
iterations = 10
steps_1param = []

for step in range(iterations + 1):
    y_pred = theta_1 * X_val
    cost = np.sum((y_pred - y_val)**2) / (2 * m)
    gradient = np.sum((y_pred - y_val) * X_val) / m
    steps_1param.append([step, theta_1, cost, gradient])
    theta_1 -= alpha * gradient

df_1 = pd.DataFrame(steps_1param, columns=["Step", "Theta_1", "Cost J", "Gradient"])
print("--- 1-Parameter GD (Eq. 1.1) ---")
print(df_1.to_string(index=False))
print("\n" + "="*55 + "\n")

# --- B. 2-Parameter GD: y = theta_0 + theta_1*x (Eq. 1.2) ---
# Gradients from Eqs. 2.11-2.12; simultaneous update from Eq. 2.7
theta_0 = 0.0
theta_1 = 0.0
steps_2param = []

for step in range(iterations + 1):
    y_pred = theta_0 + theta_1 * X_val
    residuals = y_pred - y_val
    cost = np.sum(residuals**2) / (2 * m)
    grad_0 = np.sum(residuals) / m          # Eq. 2.11
    grad_1 = np.sum(residuals * X_val) / m  # Eq. 2.12
    steps_2param.append([step, theta_0, theta_1, cost, grad_0, grad_1])
    theta_0 -= alpha * grad_0  # Eq. 2.7
    theta_1 -= alpha * grad_1  # Eq. 2.7

df_2 = pd.DataFrame(steps_2param, columns=["Step", "Theta_0", "Theta_1", "Cost J", "Grad_0", "Grad_1"])
print("--- 2-Parameter GD (Eq. 1.2) ---")
print(df_2.to_string(index=False))

# --- Convergence Plots ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Cost vs iteration
axes[0].plot(df_2["Step"], df_2["Cost J"], marker='o', color='blue')
axes[0].set_title("Cost J(θ) vs Iteration (Eq. 1.6, 2-Param GD)")
axes[0].set_xlabel("Iteration Step")
axes[0].set_ylabel("Cost J")
axes[0].grid(True)

# Plot 2: Evolution of the regression line (Eq. 1.2)
axes[1].scatter(X_val, y_val, color='red', s=80, label='Data', zorder=5)
x_line = np.linspace(0, 4, 10)
for step in [0, 1, 2, 5, 10]:
    t0 = df_2.loc[step, "Theta_0"]
    t1 = df_2.loc[step, "Theta_1"]
    axes[1].plot(x_line, t0 + t1 * x_line, label=f"Step {step}")
axes[1].set_title("Regression Line Evolution (Eq. 1.2)")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
