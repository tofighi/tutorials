## Granular Taxonomy of AI/ML Algorithms

AI/ML algorithms can be categorized in a more granular way beyond the basic supervised, unsupervised, and reinforcement learning. This involves looking at the specific techniques and their underlying mathematical principles.

### Definition
A granular taxonomy classifies AI/ML algorithms by their specific approach, mathematical foundations, and problem-solving methodologies, allowing for a more precise understanding of their capabilities and limitations.

### Example
Instead of just 'Supervised Learning', a granular taxonomy might include 'Decision Trees', 'Support Vector Machines (SVMs)', 'Neural Networks (specifically Feedforward Networks, Recurrent Neural Networks, Convolutional Neural Networks)', 'Gradient Boosting Machines (e.g., XGBoost, LightGBM)', and 'Logistic Regression'. Each of these has distinct characteristics and is suited for different types of problems, even within the same overarching category.

<a name='Linear_Nonlinear'></a>
## Linear/Non-linear AI/ML

This distinction refers to the nature of the relationship an algorithm models between input features and the output target.

### Definition
*   **Linear AI/ML Models:** These models assume a linear relationship between the input features and the target variable. The output is a linear combination of the inputs.
*   **Non-linear AI/ML Models:** These models can capture complex, non-linear relationships between input features and the target variable, allowing for more flexible and intricate decision boundaries or prediction functions.

### Example
*   **Linear Model Example:** **Linear Regression** is a classic example. If you're predicting house prices based on size, a linear model might assume that for every extra square foot, the price increases by a fixed amount. The decision boundary for classification is a straight line or a hyperplane.
*   **Non-linear Model Example:** **Decision Trees**, **Random Forests**, and **Neural Networks** are typically non-linear. A decision tree can split the data space with horizontal and vertical lines, creating a piecewise linear, but overall non-linear, decision boundary. Neural networks, with their activation functions, can model highly complex, curved relationships.

<a name='Convex_Nonconvex'></a>
## Convex/Non-convex AI/ML

This concept relates to the optimization problem that an AI/ML model tries to solve during training, specifically concerning the shape of its loss function.

### Definition
*   **Convex Optimization Problems:** In a convex problem, the loss function has a single global minimum. This means that any local minimum found by an optimization algorithm (like Gradient Descent) is also the global minimum, ensuring that the algorithm converges to the best possible solution.
*   **Non-convex Optimization Problems:** In a non-convex problem, the loss function can have multiple local minima, saddle points, and plateaus. Optimization algorithms are not guaranteed to find the global minimum and might get stuck in a local minimum, leading to suboptimal model performance.

### Example
*   **Convex Example:** **Linear Regression with Mean Squared Error (MSE)** loss function or **Logistic Regression** (when optimizing with L2 regularization) typically result in convex optimization problems. This is why these models are relatively easy to train and less prone to hyperparameter tuning issues related to convergence.
*   **Non-convex Example:** The loss functions for **deep Neural Networks** are almost always non-convex. This is a major reason why training deep learning models can be challenging, requiring sophisticated optimization techniques (e.g., Adam, RMSprop) and careful initialization to navigate the complex loss landscape and find good solutions.

<a name='Training_Workflow'></a>
## Workflow for Training/Validating/Testing AI/ML Models

A structured approach is crucial for building robust and reliable AI/ML models. This workflow ensures that the model generalizes well to unseen data and avoids overfitting.

### Definition
The workflow for training, validating, and testing AI/ML models typically involves splitting the available dataset into three distinct subsets: a training set (used to train the model), a validation set (used to tune hyperparameters and prevent overfitting during development), and a test set (used for a final, unbiased evaluation of the model's performance on completely new data).

### Example
Consider building a model to predict customer churn:
1.  **Data Collection:** Gather historical customer data, including features like usage, billing, and demographics, along with a 'churned' label.
2.  **Data Split:** Divide the dataset into:
    *   **Training Set (e.g., 70%):** Used to teach the model the patterns in the data.
    *   **Validation Set (e.g., 15%):** Used during model development to evaluate different algorithms, fine-tune hyperparameters (e.g., learning rate, number of layers in a neural network), and decide when to stop training to prevent overfitting. This set helps select the best model configuration.
    *   **Test Set (e.g., 15%):** Held out until the very end. Once the model is fully developed and tuned using the training and validation sets, the test set provides a final, unbiased measure of how the model will perform on new, unseen customers. It simulates real-world performance.

## Methodology for Feature Engineering

Feature engineering is a critical step in the machine learning pipeline, often having a greater impact on model performance than the choice of algorithm itself.

### Definition
Feature engineering is the process of using domain knowledge to extract new features from raw data that help machine learning models perform better. It involves transforming existing features or creating new ones to improve the model's ability to learn patterns and make accurate predictions.

### Example
Imagine you have a dataset of customer transactions, including a `timestamp` column and a `product_category` column.

Raw features:
*   `timestamp`: `2023-10-26 14:35:00`
*   `product_category`: `Electronics`

Engineered features from `timestamp`:
*   `hour_of_day`: `14` (Is a purchase at 2 PM more likely for certain categories?)
*   `day_of_week`: `Thursday` (Are weekend purchases different from weekday purchases?)
*   `is_weekend`: `True/False`
*   `month`: `10` (Are there seasonal purchasing patterns?)

Engineered features from `product_category` and other data:
*   `has_purchased_luxury_item`: `True/False` (Derived from product categories, indicating customer segment)
*   `transaction_count_last_30_days`: (Aggregating historical transaction data)

These new features can provide the model with more meaningful information than the raw `timestamp` and `product_category` alone, allowing it to capture more nuanced patterns.

<a name='Data_Labelling'></a>
## Data Labelling

Data labelling is the process of identifying raw data (images, text files, videos, etc.) and adding one or more informative labels to provide context. This is fundamental for supervised learning.

### Definition
Data labelling (or data annotation) is the process of tagging or classifying data with meaningful attributes (labels) that a machine learning model can learn from. For supervised learning, high-quality labeled data is essential for the model to correctly identify patterns and make accurate predictions.

### Example
*   **Image Classification:** For an image recognition model that distinguishes between cats and dogs, you would manually (or semi-automatically) label thousands of images with either 'cat' or 'dog'. For object detection, you might draw bounding boxes around each cat and dog and label them.
*   **Natural Language Processing (NLP):** For a sentiment analysis model, sentences like 'This movie was great!' would be labeled 'positive', and 'I hated the ending.' would be labeled 'negative'. For named entity recognition, specific words in text would be tagged as 'person', 'organization', 'location', etc.
*   **Medical Imaging:** For a model detecting tumors in X-rays, radiologists would outline and label cancerous regions in medical scans.

The accuracy and consistency of these labels directly impact the performance and reliability of the trained AI/ML model.

# Artificial Intelligence (AI) vs. Machine Learning (ML) vs. Data Science

These terms are often used interchangeably, but they represent distinct, though overlapping, fields.

### Artificial Intelligence (AI)

*   **Definition:** AI is the broadest field, encompassing any technique that enables computers to mimic human intelligence. The goal of AI is to create intelligent agents that can reason, learn, perceive, understand, and interact.
*   **Scope:** It includes traditional AI (rule-based systems, expert systems), machine learning, deep learning, natural language processing, computer vision, robotics, and more.
*   **Analogy:** Think of AI as the entire universe of making machines intelligent, where the goal is to make computers 'think' like humans or perform tasks that typically require human intelligence.
*   **Example:** A self-driving car (combining computer vision, decision-making logic, control systems), a chess-playing computer (using search algorithms and heuristics), or a natural language understanding system that can hold a conversation.

### Machine Learning (ML)

*   **Definition:** ML is a subset of AI that focuses on building systems that can learn from data without explicit programming. Instead of being explicitly told how to solve a problem, ML algorithms learn patterns from data and use those patterns to make predictions or decisions.
*   **Scope:** Supervised learning, unsupervised learning, reinforcement learning, deep learning (a subfield of ML).
*   **Analogy:** If AI is the universe, then ML is a galaxy within that universe – a powerful approach to achieve AI by learning from data.
*   **Example:** A spam filter that learns to identify spam emails based on examples, a recommendation system that suggests products based on past purchases, or a model that predicts house prices based on various features.

### Data Science

*   **Definition:** Data Science is an interdisciplinary field that uses scientific methods, processes, algorithms, and systems to extract knowledge and insights from structured and unstructured data. It combines elements of statistics, computer science, domain expertise, and data visualization.
*   **Scope:** Data collection, data cleaning, data exploration (EDA), data visualization, statistical modeling, machine learning, communication of insights, and deployment of data-driven solutions.
*   **Analogy:** Data Science is like a large city within the AI universe, where practitioners use various tools (including ML algorithms) to understand data and solve real-world problems.
*   **Example:** Analyzing sales data to understand customer behavior and identify growth opportunities, predicting disease outbreaks using public health data, or optimizing logistics routes using traffic and weather data. While Data Scientists often use ML, their role is broader, encompassing the entire lifecycle of data analysis and insight generation.

## What is Machine Learning?

Machine Learning (ML) is the scientific study of algorithms and statistical models that computer systems use to effectively perform a specific task without using explicit instructions, relying on patterns and inference instead. Machine Learning is considered as a subset of Artificial Intelligence (AI). Machine learning algorithms build a mathematical model of sample data, known as *training data*, in order to make predictions or decisions without being explicitly programmed to perform the task.

Arthur Samuel in 1959 defined Machine Learning as a "*field of study that gives computers the ability to learn without being explicitly programmed*." (Géron, 2017)

Another famous and more engineering-oriented definition is presented by Tom Mitchel in 1997: "A computer program is said to learn from experience *E* with respect to some task *T* and some performance measure *P*, if its performance on T, as measured by P, improves with experience E."(Géron, 2017)

### Example - Spam Filter

Advanced spam filters are using Machine Learning programs which can learn to flag
spam given examples of spam emails (e.g., flagged by users) and examples of regular
(non-spam, also called “ham”) emails. The examples that the system uses to learn are
called the training set which is shown in the figure below. Each training example is called a training **instance** (or sample).
In this case, the task T is to flag spam for new emails, the experience E is the training
data, and the performance measure P needs to be defined; for example, you can use
the ratio of correctly classified emails. This particular performance measure is called
accuracy and it is often used in classification tasks.

**NOTE:** Training a Machine Learning algorithm is not storing huge data in computers. For example, if you just download a copy of Wikipedia, your computer has a lot more data, but it is not suddenly better at any task. Thus, it is not Machine Learning.

## Instance-Based versus Model-Based Learning

One more way to categorize Machine Learning systems is by how they generalize. Most Machine Learning tasks are about making predictions. This means that given a number of training examples, the system needs to be able to generalize to examples it has never seen before. Having a good performance measure on the training data is good, but insuﬃcient; the true goal is to perform well on new instances. There are two main approaches to generalization: instance-based learning and model-based learning.

<a name='Instance_Based_Learning'></a>
### Instance-based learning

Possibly the most trivial form of learning is simply to learn by heart. If you were to
create a spam filter this way, it would just flag all emails that are identical to emails
that have already been flagged by users—not the worst solution, but certainly not the
best.

Instead of just flagging emails that are identical to known spam emails, your spam
filter could be programmed to also flag emails that are very similar to known spam
emails. This requires a measure of similarity between two emails. A (very basic) similarity
measure between two emails could be to count the number of words they have
in common. The system would flag an email as spam if it has many words in common
with a known spam email.

This is called instance-based learning: the system learns the examples by heart, then
generalizes to new cases using a similarity measure (figure below). In the figure below, **k-Nearest Neighbours (k-NN)** is shown as an example of Instance based learning.  The training samples (instances) are shown in the figure below.  We have two features (feature 1 and feature 2) and two classes (Triangles and Squares). Each of the data points (training instances) are labeled as Triangle or Square.

**X** is a new data point which does not exist in the dataset and its label is unknown. Our task here is finding the suitable label to assign to this new data point (Triangle or Square). This is a classification task, and the algorithm tries to find the closest training examples to this new data point to see which class label is more suitable. Therefore, all the training examples should be stored in the algorithm. The similarity measure here is calculated based on the distance of new data points to all training instances. If we choose $k=3$, it finds 3 closest neighbours and count how many of them are Triangles and how many are Squares. The majority class will be assigned which in this example is Triangle. Therefore, the class of new data point should be Triangle.

<a name='Building_Blocks_Model_Based_ML'></a>
## Building Blocks of a Model-based Machine Learning Algorithm

### 1) A hypothesis (model) from X (feature) to y (target)
**Example:** Linear Regression Model for each sample:
$$\hat{y} = h(x) = \sum_{i=0}^{d}\theta_{i}x_{i} = \theta^{T}x$$

### 2) A loss function
**Example:** Linear Regression Loss for each sample:
$$\text{Loss} = (\hat{y} - y)^{2} = (h(x) - y)^{2}$$

### 3) An optimization criterion based on the loss function (a cost function, for example)
**Example:** Linear Regression cost function for all training samples:
$$J(\theta) = \frac{1}{2m}\sum_{i=1}^{m}(h(x^{i}) - y^{i})^{2}$$

### 4) An optimization routine leveraging training data to find a solution to the optimization criterion
**Example:** Gradient Descent or Normal Equations for Linear Regression to find the best $\theta$.

## Feature Matrix (X) and Target Variable (y)

Data preprocessing modules transform raw data to produce the **Feature Matrix ($X$)** and **Target Vector ($y$)**:
- **Feature Matrix ($X$):** Structured as $n\_samples \times n\_features$.
- **Target Vector ($y$):** Structured as $n\_samples \times 1$.

Each feature is finally presented by one of the following:
- **Numerical data:** Infinite number of values (e.g., age, weight, ...)
- **Categorical data:** Finite number of values (e.g., gender, hair color, ...)

## Standardization (Normalization)

In machine learning, we can handle various types of data, e.g., audio signals and pixel values for image data, and this data can include multiple dimensions.

Feature standardization makes the values of each feature in the data have **zero-mean** (when subtracting the mean in the numerator) and **unit-variance**. This method is widely used for normalization in many machine learning algorithms.

For each column (feature) in $X$, we need to transform:

$$x' = \frac{x - \bar{x}}{\sigma}$$

Where:
- $x$ is the original feature vector
- $\bar{x} = \text{average}(x)$ is the mean of that feature vector
- $\sigma$ is its standard deviation

## Machine Learning Starts Here

**A1.** This is a **Classification** task (a branch of Supervised Learning).

**A3.** Convex problems have a single global minimum, ensuring optimization algorithms reach the best solution. Non-convex problems have multiple local minima where an optimizer might get stuck.

**A4.** The subsets are **Training, Validation, and Test**. The Validation Set is used to tune hyperparameters and select the best model configuration before final testing.

**A8.** 1. Hypothesis (Model), 2. Loss Function, 3. Optimization Criterion (Cost Function), 4. Optimization Routine.

**A9.** **Unsupervised Learning** (specifically, Clustering).
